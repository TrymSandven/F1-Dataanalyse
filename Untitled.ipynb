```python
from matplotlib import pyplot as plt
import fastf1
import fastf1.plotting
```

```python
fastf1.plotting.setup_mpl()

session = fastf1.get_session(2022, 'Barcelona', 'FP2')

session.load()
fast_leclerc = session.laps.pick_driver('VET').pick_fastest()
```

```python
ver_lap = session.laps.pick_driver('VER').pick_fastest()
vet_lap = session.laps.pick_driver('VET').pick_fastest()
```

```python
ver_tel = ver_lap.get_car_data().add_distance()
vet_tel = vet_lap.get_car_data().add_distance()
```

```python
rbr_color = fastf1.plotting.team_color('RBR')
amr_color = fastf1.plotting.team_color('AMR')

fig, ax = plt.subplots()
ax.plot(ver_tel['Distance'], ver_tel['Speed'], color=rbr_color, label='VER')
ax.plot(vet_tel['Distance'], vet_tel['Speed'], color=amr_color, label='VET')

ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} FP2")

plt.show()
```
